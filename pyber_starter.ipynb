```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
#df['count'] = df.groupby('group')['group'].transform('count')
# Combine the data into a single dataset
ride_data['count'] = ride_data.groupby(["city"])["city"].transform("count")
ride_data = ride_data.groupby(['city']).mean()
df = pd.DataFrame(data = (city_data.merge(ride_data,on = 'city', how = 'left')))
# Display the data table for preview
df
```

```python
urban_df = pd.DataFrame(data = (df.loc[df['type'] == 'Urban']))
rural_df = pd.DataFrame(data = (df.loc[df['type'] == 'Rural']))
suburban_df = pd.DataFrame(data = (df.loc[df['type'] == 'Suburban']))
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter pblots for each city types
teststr = 'Note: \nCircle size correlated with number of drivers in city'
plt.scatter(urban_df['count'], urban_df['fare'], s = urban_df['driver_count'] * 10, c = 'red', alpha = .5 , label = 'Urban', edgecolors = 'black')
plt.scatter(rural_df['count'], rural_df['fare'], s = rural_df['driver_count'] * 10, c = 'gold', alpha = .5, label = 'Rural', edgecolors = 'black' )
plt.scatter(suburban_df['count'], suburban_df['fare'], s = suburban_df['driver_count'] * 10, c = 'blue', alpha = .5, label = 'Suburban', edgecolors = 'black' )
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides(Per City)')
plt.ylabel('Average Fare ($)')
plt.text(42,30,teststr)
plt.grid()
plt.legend(title = 'City Types')
plt.show()

# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python

total = df['fare'].count()
urbanper = round((((urban_df['fare'].count()) / total)*100) , 1)
ruralper = round((((rural_df['fare'].count()) / total)*100) , 1)
suburbanper = round((((suburban_df['fare'].count()) / total)*100) , 1)
percentage = [urbanper, ruralper, suburbanper]
labels = ['Urban', 'Rural', 'Suburban']
color = ['red' , 'blue' , 'yellow']
explode = [.1,0,0]
plt.pie(percentage,
        explode = explode,
        labels = labels, 
        colors = color,
        autopct = '%1.1f%%',
        shadow = True ,
        startangle = -45)
plt.title('% of Total Fares by City Type')
plt.show
```

## Total Rides by City Type

```python

total = df['count'].sum()
urbanrid = round((((urban_df['count'].sum()) / total)*100) , 1)
ruralrid = round((((rural_df['count'].sum()) / total)*100) , 1)
suburbanrid = round((((suburban_df['count'].sum()) / total)*100) , 1)
percentage = [urbanrid, ruralrid, suburbanrid]
labels = ['Urban', 'Rural', 'Suburban']
color = ['red' , 'blue' , 'yellow']
explode = [.1,0,0]
plt.pie(percentage,
        explode = explode,
        labels = labels, 
        colors = color,
        autopct = '%1.1f%%',
        shadow = True ,
        startangle = -70)
plt.title('% of Total Rides by City Type')
plt.show
```

## Total Drivers by City Type

```python

total = df['driver_count'].sum()
urbandrv = round((((urban_df['driver_count'].sum()) / total)*100) , 1)
ruraldrv = round((((rural_df['driver_count'].sum()) / total)*100) , 1)
suburbandrv = round((((suburban_df['fare'].sum()) / total)*100) , 1)
percentage = [urbandrv, ruraldrv, suburbandrv]
labels = ['Urban', 'Rural', 'Suburban']
color = ['red' , 'blue' , 'yellow']
explode = [.1,0,0]
plt.pie(percentage,
        explode = explode,
        labels = labels, 
        colors = color,
        autopct = '%1.1f%%',
        shadow = True ,
        startangle = -70)
plt.title('% of Total Fares by City Type')
plt.show
```
